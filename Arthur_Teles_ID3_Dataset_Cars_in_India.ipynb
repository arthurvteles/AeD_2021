Calcula a entropia de uma determinada classe

```python
def entropy(df,feature_label,feature_value ,target_label ):
  entropy = 0

  if feature_label == target_label:
    dataset_size = df.shape[0]
    tv_count = df[target_label].value_counts()[feature_value]
    probability = tv_count/dataset_size
    entropy_tv = - probability * np.log2(probability)
    entropy += entropy_tv
    return entropy


  new_dataset = df.loc[(df[feature_label] == feature_value)]
  new_dataset = new_dataset[[feature_label, target_label]]
  new_dataset_size = new_dataset[feature_label].count()
  target_value = new_dataset[target_label].unique()


  for tv in target_value:
    tv_count = new_dataset[target_label].value_counts()[tv]
    probability = tv_count/new_dataset_size
    entropy_tv = - probability * np.log2(probability)
    entropy += entropy_tv

  return entropy


```

Calcula o ganho de informação

```python
def info_gain(df, feature_label, target_label, total_entropy):
  feature_label_values = df[feature_label].unique()
  df_size = df.shape[0]
  gain = 0.0

  for flv in feature_label_values:
    feature_entropy = entropy(df,feature_label, flv, target_label)
    feature_count = df[df[feature_label] == flv].shape[0]
    feature_probability = feature_count/df_size
    gain += feature_probability * feature_entropy

  return total_entropy - gain
```

```python
#calcular a entropia total
target_values = test['NCAP Rating'].unique()
total_entropy = 0

for tv in target_values:
  tv_entropy = entropy(test,'NCAP Rating', tv, 'NCAP Rating')
  total_entropy += tv_entropy
```

Finding the most informative feature

```python
def finding_the_most_informative(df, target_label, total_entropy):
  most_informative = {}
  features = df.columns.drop(target_label)

  for ft in features:
    feature_gain = info_gain(df,ft,target_label,total_entropy)
    new_data = {ft:feature_gain}
    most_informative.update(new_data)

  most_informative = sorted(most_informative.items(), key= lambda x:x[1], reverse = True )
  return most_informative

print(finding_the_most_informative(test,'NCAP Rating', total_entropy))
```
